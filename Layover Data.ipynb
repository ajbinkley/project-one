```python
# Gather purchased ticket data (eceonomy class) from the last 48 hours
hubs=["ATL","NYC","HKG","CDG","GIG","JNB"]
# leaving [Atlanta, New York City, Hong Kong, Paris, Rio de Janiero, Johannesburg]

# Desired data
prices=[]
origins=[]
destinations=[]
distances=[]
departs=[]
returns=[]
purchases=[]
cheapests=[]
transfers=[]


# Request up to 500 flights from each hub 
for hub in hubs:
    prices_url=f"http://api.travelpayouts.com/v2/prices/latest?currency=usd&limit=500&origin={hub}"
    headers = {"X-Access-Token": "f2b1a8fac25474496ea091529cddd0ec",
             }
    prices_data = requests.get(prices_url, headers=headers).json()

    for i in prices_data["data"]:
        destination=i["destination"]
        depart=i["depart_date"]
        returnon=i["return_date"]
#       Pull data for cheapest flight of the same parameters
        month_data = requests.get(f"http://api.travelpayouts.com/v1/prices/calendar?currency=usd&origin={hub}&destination={destination}&depart_date={depart}&return_at={returnon}", headers=headers).json()
        if len(month_data["data"])>0:
            day_data=month_data["data"][returnon]
            cheapest=day_data["price"]
            transfer=day_data["transfers"]
            cheapests.append(cheapest)
            transfers.append(transfer)
            prices.append(i["value"])
            origins.append(hub)
            destinations.append(destination)
            distances.append(int(int(i["distance"])//1.609)) #convert km to miles
            departs.append(depart)
            purchases.append(i["found_at"])
            returns.append(returnon)

# Store data into dataframe 
prices_df=pd.DataFrame({"Origin":origins,
                        "Destination":destinations,
                        "Distance (Miles)":distances,
                        "Departure Date":departs,
                        "Return Date":returns,
                        "Price (USD)":prices,
                        "Cheapest Price":cheapests,
                        "Layovers":transfers,
                        "Purchase Time":purchases
                       })
prices_df
```
