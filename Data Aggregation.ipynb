```python
import requests
import json
import pandas as pd
import csv
from pprint import pprint
```

```python
csvpath="Resources/busiest_us_airport_codes.csv"
busyairportdf=pd.read_csv(csvpath)
busyairportdf
```

```python

airports=list(busyairportdf["Code"])

prices=[]
origins=[]
destinations=[]
distances=[]
departs=[]
purchases=[]
classes=[]
gates=[]


for origin in airports:
    print(f"Searching flights out of {origin}")
    prices_url=f"http://api.travelpayouts.com/v2/prices/latest?currency=usd&limit=1000&origin={origin}&one_way=true"
    headers = {"X-Access-Token": "f2b1a8fac25474496ea091529cddd0ec",
             }
    prices_data = requests.get(prices_url, headers=headers).json()

    for i in prices_data["data"]:
        destination=i["destination"]
        depart=i["depart_date"]
        print(f"From {origin} to {destination}.")
        prices.append(i["value"])
        origins.append(origin)
        destinations.append(destination)
        distances.append(int(int(i["distance"])//1.609)) #convert km to miles
        departs.append(depart)
        purchases.append(i["found_at"])
        classes.append(i["trip_class"])
        gates.append(i["gate"])

# Store data into dataframe 
prices_df=pd.DataFrame({"Origin":origins,
                        "Destination":destinations,
                        "Purchase Time":purchases,
                        "Distance (Miles)":distances,
                        "Departure Date":departs,
                        "Price (USD)":prices,
                        "Class":classes,
                        "Purchase Gate":gates
                       })
```
