```python
import requests
import json
import pandas as pd
from pprint import pprint
```

```python
# Gather purchased ticket data (eceonomy class) from the last 48 hours
hubs=["ATL","NYC","HKG","CDG","GIG","JNB"]
# leaving [Atlanta, New York City, Hong Kong, Paris, Rio de Janiero, Johannesburg]

# Desired data
prices=[]
origins=[]
destinations=[]
distances=[]
departs=[]
returns=[]
purchases=[]
directs=[]
cheapests=[]
# transfers=[]


# Request up to 500 flights from each hub 
for hub in hubs:
    print(f"Searching flights out of {hub}")
    prices_url=f"http://api.travelpayouts.com/v2/prices/latest?currency=usd&limit=500&origin={hub}"
    headers = {"X-Access-Token": "f2b1a8fac25474496ea091529cddd0ec",
             }
    prices_data = requests.get(prices_url, headers=headers).json()

    for i in prices_data["data"]:
        destination=i["destination"]
        depart=i["depart_date"]
        returnon=i["return_date"]
#       Pull data for cheapest direct flight of the same parameters
        direct_data = requests.get(f"http://api.travelpayouts.com/v1/prices/direct?currency=USD&origin={hub}&destination={destination}&depart_date={depart}&return_date={returnon}", headers=headers).json()
        try:
            direct=direct_data["data"][destination]["0"]["price"]
        except:
            direct=float('nan')    
        directs.append(direct)
#       Pull data for cheapest flight (allowing layovers) of the same parameters
        cheap_data = requests.get(f"http://api.travelpayouts.com/v1/prices/cheap?currency=USD&origin={hub}&destination={destination}&depart_date={depart}&return_date={returnon}", headers=headers).json()
        try:
            cheapest=cheap_data["data"][destination]["0"]["price"]
        except:
            cheapest=float('nan')
        print(f"From {hub} to {destination}.")
        cheapests.append(cheapest)
        prices.append(i["value"])
        origins.append(hub)
        destinations.append(destination)
        distances.append(int(int(i["distance"])//1.609)) #convert km to miles
        departs.append(depart)
        purchases.append(i["found_at"])
        returns.append(returnon)

# Store data into dataframe 
prices_df=pd.DataFrame({"Origin":origins,
                        "Destination":destinations,
                        "Distance (Miles)":distances,
                        "Departure Date":departs,
                        "Return Date":returns,
                        "Price (USD)":prices,
                        "Cheapest Direct Price":directs,
                        "Cheapest Price":cheapests,
#                         "Layovers":transfers,
                        "Purchase Time":purchases
                       })
```

```python
prices_df["Layover Savings"]=prices_df["Cheapest Direct Price"]-prices_df["Cheapest Price"]

```

```python
savings=prices_df["Layover Savings"].value_counts(ascending=True, dropna=True)
savings
```
